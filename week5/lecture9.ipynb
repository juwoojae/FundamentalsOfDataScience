# Mesasures of Central Tendency
### 1. Mean
> 이상치(outliers)에 민감하고, 대칭이면서, 단봉형일때 적합하다

### 2. Median
> 이상치 커버에 강하고, Skewed 분포에 강하다.

```python
import numpy as np

nums = np.array([10, 12, 14, 16, 18])
print('mean:', np.mean(nums))
print('median:', np.median(nums))

nums = np.concat((nums, [999])) # concat 은 튜플이 들어가야 한다.
print('mean:', np.mean(nums))
print('median:', np.median(nums)) # 중앙 값은 outlier 일때 효과적이다.
```

### Correlation analysis using NumPy

```python
import numpy as np

x = np.arange(0,10) # 0 부터 9까지의 정수 배열 생성
y1 = x * 2 +1;

np.corrcoef(x,y1) # 상관 계수를 계산하는 코드
# 완벽한 양의 선형관계라서 1 이다.

np.random.seed(2) # 난수 생성 고정 (재현 가능하게)

x = np.arange(0,10) # 0 부터 9까지의 배열 생성
y2 = np.random.randint(0,100, size=10) # 0~100 사이 랜덤 정수 10개 생성
np.corrcoef(x,y2) # 상관 계수 계산

np.corrcoef((x,y1,y2))
```
