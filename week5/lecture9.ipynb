# Mesasures of Central Tendency
### 1. Mean
> 이상치(outliers)에 민감하고, 대칭이면서, 단봉형일때 적합하다

### 2. Median
> 이상치 커버에 강하고, Skewed 분포에 강하다.

```python
import numpy as np

nums = np.array([10, 12, 14, 16, 18])
print('mean:', np.mean(nums))
print('median:', np.median(nums))

nums = np.concat((nums, [999])) # concat 은 튜플이 들어가야 한다.
print('mean:', np.mean(nums))
print('median:', np.median(nums)) # 중앙 값은 outlier 일때 효과적이다.
```

### Correlation analysis using NumPy

```python
import numpy as np

x = np.arange(0,10) # 0 부터 9까지의 정수 배열 생성
y1 = x * 2 +1;

np.corrcoef(x,y1) # 상관 계수를 계산하는 코드
# 완벽한 양의 선형관계라서 1 이다.

np.random.seed(2) # 난수 생성 고정 (재현 가능하게)

x = np.arange(0,10) # 0 부터 9까지의 배열 생성
y2 = np.random.randint(0,100, size=10) # 0~100 사이 랜덤 정수 10개 생성
np.corrcoef(x,y2) # 상관 계수 계산

print(np.corrcoef((x,y1,y2)))
```

# Seaborn library

```python
# 필요한 라이브러리 import
import seaborn as sns               # 시각화 라이브러리
import matplotlib.pyplot as plt     # 그래프 출력용

# 예제 데이터셋 로드 (Seaborn 기본 제공)
df = sns.load_dataset("tips")       # 레스토랑 팁 데이터
print(df)
# -----------------------------
# 1. 히스토그램 (분포 확인)
# -----------------------------
sns.histplot(df['total_bill'], bins=10)   # total_bill 분포
plt.title("Histogram of Total Bill")      # 제목 설정
plt.show()                                # 그래프 출력

# -----------------------------
# 2. 박스플롯 (이상치 + 분포)
# -----------------------------
sns.boxplot(x=df['total_bill'])           # 박스플롯
plt.title("Boxplot of Total Bill")
plt.show()

# -----------------------------
# 3. 바이올린 플롯 (분포 + 밀도)
# -----------------------------
sns.violinplot(x=df['total_bill'])        # 바이올린 플롯
plt.title("Violin Plot of Total Bill")
plt.show()

# -----------------------------
# 4. 막대그래프 (카테고리 비교)
# -----------------------------
sns.barplot(x='day', y='total_bill', data=df)   # 요일별 평균 total_bill
plt.title("Average Total Bill by Day")
plt.show()

# -----------------------------
# 5. 산점도 (관계 확인)
# -----------------------------
sns.scatterplot(x='total_bill', y='tip', data=df)
plt.title("Scatter Plot (Total Bill vs Tip)")
plt.show()

# -----------------------------
# 6. 회귀선 포함 산점도
# -----------------------------
sns.regplot(
    x='total_bill',
    y='tip',
    data=df,
    scatter_kws={"s":100, "alpha":0.3},   # 점 스타일
    line_kws={"lw":3, "ls":"--", "alpha":0.5}  # 선 스타일
)
plt.title("Regression Plot (Total Bill vs Tip)")
plt.show()

sns.pairplot(
    df,
    vars=['total_bill', 'tip', 'size'],  # 확인할 변수들
    hue='sex'                            # 색으로 그룹 구분
)
plt.show()

corr = df.corr(numeric_only=True)   # 상관계수 계산

sns.heatmap(
    corr,
    annot=True,     # 값 표시
    fmt=".2f",      # 소수점 2자리
    cmap="Blues"    # 색상 테마
)
plt.show()
# 출력
```

```python
import matplotlib.pyplot as plt        # matplotlib import
import seaborn as sns                 # 데이터 로드용

tips = sns.load_dataset("tips")       # 데이터 로드

# matplotlib으로 total_bill과 tip 관계를 그리는 산점도 코드
plt.scatter(tips['total_bill'], tips['tip'])   # 산점도 생성
plt.xlabel('total_bill')             # x축 이름
plt.ylabel('tip')                    # y축 이름
plt.title('Total Bill vs Tip')       # 제목 설정
plt.show()                           # 그래프 출력
```

with style, size, etc
* 스타일, 크기 등 추가 옵션 가능
* 더 많은 정보 표현 가능

```python
sns.relplot(
    x='total_bill',
    y='tip',
    hue='smoker',     # 색으로 구준
    style='smoker',   # 마커 모양 (모양으로 구분)
    size='size',      # 점 크기 (인원 수) 크기로 표현
    data=tips
)
```
