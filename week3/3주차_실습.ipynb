### 3. `plot()`의 선 스타일


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-20, 20)   # -20에서 20 사이의 수를 1의 간격으로 생성.
y1 = 2 * x               # 2*x를 원소로 가지는 y1 함수
y2 = (1/3) * x ** 2 + 5  # (1/3)*x**2 + 5를 원소로 가지는 y2 함수
y3 = -x ** 2 - 5         # -x**2 - 5를 원소로 가지는 y3 함수
# 빨강색 점선, 녹색 실선과 세모기호, 파랑색 별표와 점선으로 함수를 표현
plt.plot(x, y1, 'g--', x, y2, 'r^-', x, y3, 'b*:')
plt.axis([-30, 30, -30, 30]) # 그림을 그릴 영역을 지정함
```

### 실습: 사인 함수와 코사인 함수 그리기


```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(0, np.pi * 2, 100) #2*pi는 360도에 해당함

# Todo 1: sin 함수 만듦. np.sin() 이용
y1 = np.sin(x)

# Todo 2: cos 함수 만듦. np.cos() 이용
y2 = np.cos(x)

# Todo 3: sin 함수는 빨강 실선으로, cos 함수는 파랑 점선으로 그리기
plt.plot(x, y1, 'g')
plt.plot(x, y2, 'b:')
```

### 4. 그림의 제목과 축 라벨


```python
x = np.linspace(0, np.pi * 2, 100)

plt.title('Sin cos curve')
plt.plot(x, np.sin(x), 'r-', label='sin curve')
plt.plot(x, np.cos(x), 'b:', label='cos curve')
plt.xlabel('x value')
plt.ylabel('y value')
plt.legend()
```

제목과 범례를 직접 바꿔보기


```python
x = np.linspace(0, np.pi * 2, 100)

# TODO 1: 제목을 'Sin and Cos with Different Styles'로 바꿈.
plt.title('Sin and Cos with Different Styles')

# TODO 2: sin(x)는 초록 실선, cos(x)는 검정 점선으로 그림.
# TODO 3: label도 각각 'sin(x)', 'cos(x)'로 바꿈.
plt.plot(x, np.sin(x), 'g--', label='sin(x)')
plt.plot(x, np.cos(x), ':', label='cos(x)')

# TODO 4: x축 이름, y축 이름을 각각 'angle', 'value'로 바꿈.
plt.xlabel('angle')
plt.ylabel('value')
plt.legend()
plt.show()
```

### 산점도와 막대그래프


```python
N = 30                     # 전체 점의 개수
x = np.random.rand(N)      # 점들의 임의의 x 좌표
y = np.random.rand(N)      # 점들의 임의의 y 좌표
cmap = 'viridis' # 컬러맵
colors = np.random.rand(N) # 랜덤 숫자. 그래프에서는 컬러맵과 연관되어 랜덤 색상으로 변경 (2.명시적인 색, string 의 리스트로 집어넣는 방법)
area = (30 * np.random.rand(N))**2 # 점의 면적 크기를 랜덤하게 정함

plt.scatter(x, y, s=area, c=colors, cmap=cmap, alpha=0.5) # alpah 는 투명도
```
