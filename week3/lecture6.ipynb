```python
import matplotlib.pyplot as plt

x = [0, 1, 2, 3] # x 축 값 리스트
y = [0,1,4,9] # y 축 값 리스트 (x^2 형태)
plt.plot(x, y) # (x,y) 좌표를 선으로 연결해서 그래프 생성
plt.show() # 그래프를 화면에 출력하기
```

```python
# broadcasting 사용하기
import numpy as np
x = np.arange(10) # 0에서 9 사이의 정수값을 생성함
plt.plot(x, x**2) # x^2 함수를 plotting
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(-20,20)
y1 = x*2
y2 = (1/3)*x**2 +5
y3 = -x**2 -5
y4 = np.sin(x)

plt.plot(x,y1,'r:')
plt.plot(x,y2,'b>-')
plt.plot(x,y3,'g--')
plt.plot(x,y4,'y*:')
plt.axis([-30,30,-30,30])
```

```python
x = np.arange(-20,20)
y1 = x*2
y2 = (1/3)*x**2 +5
y3 = -x**2 -5
y4 = np.sin(x)

plt.plot(x,y2,'b>-')
plt.axis([-30, 30, -30, 30])   # x, y축 범위를 -30~30으로 고정
plt.figure()
plt.plot(x,y1,'r:')
plt.axis([-30, 30, -30, 30])   # x, y축 범위를 -30~30으로 고정
plt.figure()
plt.plot(x,y3,'g--')
plt.axis([-30, 30, -30, 30])   # x, y축 범위를 -30~30으로 고정
plt.figure()
plt.plot(x,y4,'y--')
plt.axis([-30, 30, -30, 30])   # x, y축 범위를 -30~30으로 고정
```

```python
x = np.linspace(0,np.pi,100) # 0,2ㅠ 를 100으로 나누기

plt.title('Sin cos curve') # plot 제목
plt.plot(x, np.sin(x), 'r:', label='sin curve') # 범례
plt.plot(x, np.cos(x), 'b--', label='cos curve') #
plt.plot(x, np.cos(x))
plt.xlabel('x value') # 축 이름 설정하기
plt.xlabel('y value')
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(2,2) # 2행 2열 subplot 생성하기
# 각 위치에 그래프 넣기
ax[0,0].plot([1,2,3],[1,4,9]) # 좌상단
ax[0,1].plot([1,2,3],[1,2,3]) # 우상단
ax[1,0].plot([1,2,3],[9,4,1]) # 좌하단
ax[1,1].plot([1,2,3],[3,2,1]) # 우하단
```

```python
fig, ax = plt.subplots(2,2) # 2행 2열의 그림을 그리는 공간을 만든다.
X = np.random.randn(100) # 0근처의 정규분포를 나타내는 데이터 100개 생성
Y = np.random.randn(100)
ax[0,0].scatter(X,Y) # 좌상단 산점도
X = np.arange(10) # 0에서 9사이의 연속값
Y=np.random.uniform(0,1,10) # 균등 분포. 모든 값이 나올 확률 동일
ax[0, 1].bar(X,Y) # 우상단 막대 그래프
X = np.linspace(0,10,100)
Y=np.cos(X)
ax[1,0].plot(X,Y)# 좌 하단 실선으로 함수를 그림
Z = np.random.uniform(0,1,(5,5)) # 0~1 균등 분포 5,5 행렬로 만들기
ax[1,1].imshow(Z) # 우 하단
```

```python
N = 30
x = np.random.rand(N) # 30 개의 난수 생성
y = np.random.rand(N) # 30 개의 난수 생성
colors = np.random.rand(N) # 색 랜덤으로
area = (30*np.random.rand(N))**2  # 크기 랜덤으로
# x,y 위치, s = area 점 크기, c = colors 점 색상, cmap = 'viridis', alpha 투명도
plt.scatter(x, y , s =area, c = colors, cmap = 'viridis', alpha = 0.5)
plt.show()
```
