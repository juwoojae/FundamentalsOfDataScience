### Series(시리즈)

```python
import numpy as np
import pandas as pd

se = pd.Series([1, 2, np.nan, 4])   # np.nan은 "숫자가 아님(not a number)"을 나타내는 특별한 부동소수점 값
se
```

### Series와 DataFrame(데이터프레임)

### 데이터에 열 추가하기

```python
df['total'] = df.sum(axis = 1) # df에 'total' 새 열 추가
df['mean'] = df.mean(axis=1)
print(df)
```

```python
df_new = df.sum(axis = 1)
df_new
```

```python
# DataFrame의 .sum과 .mean 메서드도 아래처럼 사용할 수 있음:
df['total'] = df[['2007', '2008', '2009', '2010', '2011']].sum(axis=1)
df['mean'] = df[['2007', '2008', '2009', '2010', '2011']].mean(axis=1)
print(df)
```

```python
df.drop('2007', inplace=True, axis=1)
print(df)  # 위에서 계산한 mean과 total 값이 그대로 남아 있음
```

```python
# 열을 삭제한 뒤 total과 mean 값을 다시 계산해야 함(자동으로 갱신되지 않음)
df['total'] = df[['2008', '2009', '2010', '2011']].sum(axis=1)
df['mean'] = df[['2008', '2009', '2010', '2011']].mean(axis=1)
print(df)
```

### Pandas를 이용한 간단한 데이터 분석

```python
import pandas as pd

path = 'https://github.com/dongupak/DataML/raw/main/csv/'
weather_file = path + 'weather.csv'

weather = pd.read_csv(weather_file, index_col = 0, encoding='CP949') # 한글 인코딩용
print(weather)
print(weather.head(3))
print('weather data shape :', weather.shape)
```

```python
print(weather.describe())  # 계산을 할수있는데이터에 대해서 컬럼별로 데이터 갯수, 평균, 표준편차, 최소/최대값 등의 정보를 보여준다.
# mean() 평균
# std 표준편차
```

```python
print('평균 분석 -----------------------------')
print(weather.mean())
print('표준편차  분석 -----------------------------')
print(weather.std())
```

### 데이터 구조를 바꾸는 pivot()

```python
import pandas as pd

df = pd.DataFrame({'상품' : ['시계', '반지', '반지', '목걸이', '팔찌'],
                   '재질' : ['금', '은', '백금', '금', '은'],
                   '가격': [500000, 20000, 350000, 300000, 60000]})
df
```

```python
new_df = df.pivot(index='상품', columns='재질', values='가격')
new_df.fillna(value=0)
#new_df #Nan 이 생기는데, 이것을 처리하기 위해 fillna 로 결측치 처리가 가능하다.
```

### 데이터프레임을 병합하는 concat()

```python
import pandas as pd

df_1 = pd.DataFrame( {'A' : ['a10', 'a11', 'a12'],
                      'B' : ['b10', 'b11', 'b12'],
                      'C' : ['c10', 'c11', 'c12']} ,
                      index = ['가', '나',  '다'] )

df_2 = pd.DataFrame( {'B' : ['b23', 'b24', 'b25'],
                      'C' : ['c23', 'c24', 'c25'],
                      'D' : ['d23', 'd24', 'd25']} ,
                      index = ['다', '라',  '마'] )
```

```python
df_1
```

```python
df_2
```

```python
df_3 = pd.concat( [df_1, df_2])
print(df_3)
```

```python
df_4 = pd.concat( [df_1, df_2], join='inner')
print(df_4)
```
