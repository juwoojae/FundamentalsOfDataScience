```python
import numpy as np

# ==============================
# 1. OR 게이트 학습 데이터
# ==============================

# OR 게이트 입력 데이터
# 첫 번째 값 1은 bias용 입력이다.
# 즉 [bias, x1, x2] 구조
X = np.array([
    [1, 0, 0],   # 0 OR 0
    [1, 0, 1],   # 0 OR 1
    [1, 1, 0],   # 1 OR 0
    [1, 1, 1]    # 1 OR 1
])

# OR 게이트 정답 데이터
# 퍼셉트론에서는 보통 출력값을 -1, +1로 둔다.
# OR 결과가 0이면 -1, OR 결과가 1이면 +1
Y = np.array([
    -1,   # 0 OR 0 = 0
     1,   # 0 OR 1 = 1
     1,   # 1 OR 0 = 1
     1    # 1 OR 1 = 1
])

# ==============================
# 2. 가중치와 학습률 설정
# ==============================

# 가중치 W를 랜덤하게 초기화
# 입력이 bias 포함 3개이므로 가중치도 3개 필요
W = np.random.randn(3)

# 학습률 eta
learning_rate = 0.1

# ==============================
# 3. 활성화 함수
# ==============================

def activation(value):
    # 퍼셉트론의 step function
    # 값이 0 이상이면 +1 출력
    # 값이 0보다 작으면 -1 출력
    if value >= 0:
        return 1
    else:
        return -1

# ==============================
# 4. 출력 계산 함수
# ==============================

def out(x):
    # W와 x의 내적 계산
    # tmp = w0*x0 + w1*x1 + w2*x2
    tmp = W.dot(x)

    # 내적 결과를 활성화 함수에 넣어서 최종 출력 계산
    return activation(tmp)

# ==============================
# 5. 학습 함수
# ==============================

def train(X, Y, epochs=10):
    # 함수 밖의 W와 learning_rate를 사용하기 위해 global 선언
    global W, learning_rate

    # epoch 반복
    # epoch 1번 = 전체 학습 데이터를 한 번 다 보는 것
    for t in range(epochs):

        print("epoch =", t)
        print("==============================")

        # 이번 epoch에서 가중치가 수정된 횟수
        adjusted = 0

        # 전체 학습 데이터를 하나씩 확인
        for i in range(len(X)):

            # 현재 입력 데이터
            x = X[i]

            # 현재 입력 데이터의 정답
            target = Y[i]

            # 현재 퍼셉트론의 예측 출력
            y_hat = out(x)

            # 정답과 출력의 차이
            error = target - y_hat

            # error가 0이 아니면 예측이 틀린 것
            if error != 0:
                adjusted += 1

            # 퍼셉트론 학습 규칙
            # W = W + learning_rate * error * x
            W = W + learning_rate * error * x

            print("입력:", x,
                  "정답:", target,
                  "출력:", y_hat,
                  "error:", error,
                  "변경된 가중치:", W)

        print()

        # 이번 epoch에서 한 번도 가중치 수정이 없었다면 학습 완료
        if adjusted == 0:
            print("학습 완료!")
            return

# ==============================
# 6. 학습 전 결과 확인
# ==============================

print("초기 가중치:", W)
print("\n학습 전 결과")
for i in range(len(X)):
    print("입력:", X[i], "출력:", out(X[i]), "정답:", Y[i])

print("\n\n학습 시작\n")

# ==============================
# 7. 학습 실행
# ==============================

train(X, Y, epochs=10)

# ==============================
# 8. 학습 후 결과 확인
# ==============================

print("\n최종 가중치:", W)
print("\n학습 후 결과")

for i in range(len(X)):
    print("입력:", X[i], "출력:", out(X[i]), "정답:", Y[i])
```
